> 문제 3. 2020_수능영어.txt파일은 2020년 수능 영어 문제지입니다. 영어 데이터만 추출하여 단어별로 분류해주세요.

```python
with open("2020_수능영어.txt", "r", encoding="utf-8") as fi:
    frequency = {}
    data = fi.read()
    data1 = re.findall(r'\b[a-z]{3,15}\b', data)
    
    for word in data1:
        count = frequency.get(word,0)
        frequency[word] = count + 1
    
    frequency_list = frequency.keys()
    
    for words in frequency_list:
        print(words, frequency[words])
```

> 문제 4. dataSet/기출100.txt파일은 수능영어가 절대평가로 바뀐 뒤 만든 수능 기출빈도 Top 100 파일입니다. '문제 1'에서 전처리한 데이터 중에서 기출100에 해당하는 단어들만 추출하세요.

```python
with open("기출100.txt", "r", encoding="utf-8") as f:
    data3 = f.read()
    data4 = re.findall(r'\b[a-z]{1,15}\b', data3)
    for i in data4:
        for j in frequency_list:
            if i == j:
                print(i, frequency[i])
```
